```python
# 차량이름에서 브랜드 추출하기
df['CarName'] = df['CarName'].apply(lambda x : x.split(' ')[0])
print(df['CarName'])
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in ['CarName', 'fueltype', 'aspiration', 'doornumber', 'carbody',
       'drivewheel', 'enginelocation', 'enginetype', 'cylindernumber',
       'fuelsystem']:
   df[col] = le.fit_transform(df[col])
```

```python
#레이블 인코딩 후 타겟변수와 피처 나누기
X = df.drop('price', axis = 1)
y = df['price']

# standard scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# 데이터 스플릿 하고 모델링하기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2)
print(len(X_train), len(X_test), len(y_train), len(y_test))
```

```python
# 요기서 부터 이어서 공부
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score

models = {
    'LinearRegression' : LinearRegression(),
    'RandomForestRegressor' : RandomForestRegressor(),
    'SVR' : SVR()
}
param_grids = {
    'LinearRegression' : {},
    'RandomForestRegressor' : {
        'n_estimators' : [50, 100, 200],
        'max_depth' : [None, 10, 20]
    },
    'SVR' : {
        'C' : [0.1, 1,10],
        'kernel' : ['linear', 'rbf'],
        'epsilon' : [0.1, 0.2]
    }
}
best_models = {}
for model_name, model in models.items():
    grid_search = GridSearchCV(model, param_grid=param_grids[model_name], cv=3)
    grid_search.fit(X_train, y_train)
    best_models[model_name] = grid_search.best_estimator_
    print(grid_search.best_params_)
    print(grid_search.best_score_)

# 학습 완료
# 모델 추론
from sklearn.metrics import root_mean_squared_error
for model in best_models:
    model = best_models[model]
    X_pred = model.predict(X_test)
    rmse = root_mean_squared_error(X_pred, y_test)
    r2 = r2_score(X_pred, y_test)
    print(model, rmse, r2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# 1. 모델 리스트 정의
models = {
    'LinearRegression': LinearRegression(),
    'RandomForest': RandomForestRegressor(),
    'SVR': SVR()
}

# 2. 하이퍼파라미터 그리드 설정
param_grids = {
    'LinearRegression': {},
    'RandomForest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20],
    },
    'SVR': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf'],
        'epsilon': [0.1, 0.2, 0.5]
    }
}

# 3. 모델 학습 및 튜닝을 위한 반복
best_models = {}
for model_name, model in models.items():
    print(f"Training {model_name}...")
    
    grid = GridSearchCV(model, param_grids[model_name], cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
    grid.fit(X_train, y_train)
    
    # 최적 모델 저장
    best_models[model_name] = grid.best_estimator_
    
    # 최적 파라미터와 성능 출력
    print(f"Best Parameters for {model_name}: {grid.best_params_}")
    print(f"Best CV MSE for {model_name}: {-grid.best_score_}")

# 4. 테스트 데이터에서 모델 성능 평가
print("\n--- Model Performance on Test Data ---")
for model_name, model in best_models.items():
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    print(f"{model_name} - RMSE: {rmse:.4f}, R2: {r2:.4f}")
    
```
